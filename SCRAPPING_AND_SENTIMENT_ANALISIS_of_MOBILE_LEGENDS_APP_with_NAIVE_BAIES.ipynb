```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

IMPORT PAKET YANG DIBUTUHKAN

```python
from types import MethodDescriptorType
from google_play_scraper import Sort, reviews
result, continuation_token = reviews(

     'com.mobile.legends',
     lang='id',
     country='id',
     sort=Sort.MOST_RELEVANT,
     count=2000,
     filter_score_with=None
)
```

```python
df_busu = pd.DataFrame(np.array(result), columns=['review'])

df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))

df_busu.head()
```

```python
len(df_busu.index)
```

```python
df_busu[['userName', 'score', 'at', 'content']].head()
```

```python
new_df = df_busu[['userName', 'score', 'at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=False)
sorted_df.head()
```

```python
my_df = sorted_df[['userName', 'score', 'at', 'content']]
```

```python
my_df = my_df[['content', 'score']]
```

```python
my_df.head()
```

PELABELAN

```python
def pelabelan(score):
  if score < 3 :
    return 'Negatif'
  elif score == 4 :
    return 'Positif'
  elif score == 5 :
    return 'Positif'
my_df['Label'] = my_df['score'].apply(pelabelan)
my_df.head(50)
```

```python
my_df.to_csv("scrapped_data.csv", index =False)
```
